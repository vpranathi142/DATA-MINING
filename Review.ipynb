```python
def add_matrices(mat_a, mat_b):
    """Add corresponding elements of two matrices."""
    return [[mat_a[i][j] + mat_b[i][j] for j in range(len(mat_a[0]))] for i in range(len(mat_a))]

def calculate_total_matrix_sum():
    """Calculate the cumulative sum of matrices X, Y, and Z."""
    # Calculate the sum of matrix X and matrix Y
    intermediate_sum = add_matrices(mat_x, mat_y)
    
    # Calculate the final sum by adding matrix Z
    total_sum = add_matrices(intermediate_sum, mat_z)
    
    return total_sum

# Define matrices X, Y, and Z
mat_x = [[3.7827, 3.3454, 3.2341], [2.2122, 3.5678, 3.9087], [1.1234, 2.8934, 5.9087]]
mat_y = [[3.1234, 3.0987, 3.1234], [2.1111, 3.2222, 3.3333], [1.0987, 1.3456, 5.1234]]
mat_z = [[3.1243, 3.0989, 3.1256], [2.6721, 3.6785, 3.9017], [1.1254, 2.8956, 5.9187]]

# Execute the function and print results
resultant_matrix = calculate_total_matrix_sum()
print("Final Matrix Result:")
for row in resultant_matrix:
    print(row)
```

```python
def merge_matrices(mat_1, mat_2):
    """Combine corresponding elements of two matrices."""
    return [[mat_1[row][col] + mat_2[row][col] for col in range(len(mat_1[0]))] for row in range(len(mat_1))]

def compute_overall_sum():
    """Compute the aggregate sum of matrices A, B, and C."""
    # Combine matrix A and matrix B
    partial_result = merge_matrices(mat_a, mat_b)
    
    # Combine the result with matrix C
    final_sum = merge_matrices(partial_result, mat_c)
    
    return final_sum

# Define matrices A, B, and C
mat_a = [[3.7827, 3.3454, 3.2341], [2.2122, 3.5678, 3.9087], [1.1234, 2.8934, 5.9087]]
mat_b = [[3.1234, 3.0987, 3.1234], [2.1111, 3.2222, 3.3333], [1.0987, 1.3456, 5.1234]]
mat_c = [[3.1243, 3.0989, 3.1256], [2.6721, 3.6785, 3.9017], [1.1254, 2.8956, 5.9187]]

# Execute the function and display results
summed_matrix = compute_overall_sum()
print("Combined Matrix Result:")
for row in summed_matrix:
    print(row)
```

```python
import itertools

# Initialize the items
items = ['A', 'B', 'C', 'D', 'E']

# Function to display combinations
def display_combinations(elements):
    print("\nGenerating Combinations:")
    for length in range(1, len(elements) + 1):
        combs = list(itertools.combinations(elements, length))
        print(f"Combinations of length {length}: {combs}")

# Function to display permutations
def display_permutations(elements):
    print("\nGenerating Permutations:")
    for length in range(1, len(elements) + 1):
        perms = list(itertools.permutations(elements, length))
        print(f"Permutations of length {length}: {perms}")

# Execute the functions
display_combinations(items)
display_permutations(items)
```

```python
import itertools

# Define the list of elements
elements = ['A', 'B', 'C', 'D', 'E']

def compute_combinations(elements):
    """
    Compute and display all combinations of varying lengths.
    """
    print("\nCombinations:")
    for length in range(1, len(elements) + 1):
        # Generate combinations of the specified length
        combos = list(itertools.combinations(elements, length))
        print(f"Combinations of length {length}:")
        for combo in combos:
            print(combo)

def compute_permutations(elements):
    """
    Compute and display all permutations of varying lengths.
    """
    print("\nPermutations:")
    for length in range(1, len(elements) + 1):
        # Generate permutations of the specified length
        perms = list(itertools.permutations(elements, length))
        print(f"Permutations of length {length}:")
        for perm in perms:
            print(perm)

# Execute functions to display combinations and permutations
compute_combinations(elements)
compute_permutations(elements)
```
